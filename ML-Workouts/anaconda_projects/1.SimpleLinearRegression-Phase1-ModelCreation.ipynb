```python
#Pandas is a python library designed specifically for data manipulation, analysis, file handling and cleaning
import pandas as pd
```

```python
#Reading CSV file using pandas libray through read_csv() function and assigning output to a variable for future refernce
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#printing the variable to know what is there in the CSV file
dataset
```

```python
#Assigning particular column values (YearsExperience) into variable 'independent';
#Use double square brackets to get the individual column values into variable
independent=dataset[["YearsExperience"]]

#Printing the individual column values 
independent
```

```python
#Assigning particular column values (Salary) into variable 'dependent';
#Use double square brackets to get the individual column values into variable
dependent=dataset[["Salary"]]

#Printing the individual column values 
dependent
```

```python
#sklearn is a open-source machine learning library for Python;
#It provides efficient procedure to split train set and test set data separately
#Using 'from', we can load specific functions (model_selection) into memory
from sklearn.model_selection import train_test_split

#X --> Input; Y--> Output
#X_train = input training set; X_test = input test set
#Y_train = output train set; Y_test = output test set
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.3, random_state=0)

#Printing input training set & output test set into the console
X_train,Y_test
```

```python
#sklearn is a open-source machine learning library for Python;
#LinearRegression is ML based algoritham to perform regression task
#Using 'from', we can load specific functions (linear_model) into memory
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

#fit( ) method substitues the input values into model
#after fit( ) execution, weight(w) and bais (b) would be calculated
regressor.fit(X_train, Y_train)
```

```python
#weight would be calculated in this ‘coef_’ variable
weight = regressor.coef_

#Printing weight into console
weight
```

```python
#Bais is nothing but origin which is stored in ‘intercept_’ variable
bais = regressor.intercept_

#Printing bais into console
bais
```

```python
y_pred = regressor.predict(X_test)
```

```python
#sklearn is a open-source machine learning library for Python;
#r2_score is nothing but 'Coefficient of determination' also called R² score, is used to evaluate the performance of a linear regression model
#Using 'from', we can load specific functions (r2_score) into memory
from sklearn.metrics import r2_score

#r2_score is a regression score function
r_score = r2_score(Y_test,y_pred)

#Printing evaluated value (Good model --> nearly to 1; bad model --> nearly to 0 )
r_score
```

```python
#pickle is a built-in standard library used for serializing and deserializing Python object structures
import pickle

#creating file name to store the model for future reference
fileName = "finalized_model_linear.sav"
```

```python
#dumping the model (regressor) into file Name to refer in other place
#wb --> write binary; sav --> file extension for saving moddel
pickle.dump(regressor, open(fileName,'wb'))
```

```python
#loading the saved model with fileName 
#rb --> read binary
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#Testing the model by giving the real time input
result = loaded_model.predict([[15]])

#Printing the final result
result
```
